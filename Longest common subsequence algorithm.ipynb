## Implementation of the longest common subsequence algorithm

```python
def LCS(x, y):
    m = len(x)
    n = len(y)
    c = [[0] * (n + 1) for i in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if x[i - 1] == y[j - 1]:
                c[i][j] = c[i - 1][j - 1] + 1
            else:
                c[i][j] = max(c[i][j - 1], c[i - 1][j])
    return c

def backtrack_c(C, X, Y, i, j):
    if i == 0 or j == 0:
        return ""
    elif X[i-1] == Y[j-1]:
        return backtrack_c(C, X, Y, i-1, j-1) + X[i-1]
    else:
        if C[i][j-1] > C[i-1][j]:
            return backtrack_c(C, X, Y, i, j-1)
        else:
            return backtrack_c(C, X, Y, i-1, j)
```
